Câu 2: Khởi tạo dataframe bằng các phương pháp khác nhau, hãy:

- Tạo dataframe có kích thước 10x3 từ cấu trúc list tương ứng, với dữ liệu cho dataframe
này gồm các cột Họ tên, Tuổi và Giới tính. Xuất kết quả ra màn hình;

```python
ho_va_ten = pd.Series(["Nguyễn Duy Tân", "Lê Bá Quốc Trung", "Lê Quang Trung", "Nguyễn Phan Hữu Phát", "Lê Gia Hân", "Trần Phúc Thái Trường", "Hoàng Ngọc Bảo Anh", "Nguyễn Thùy Trang", "Hồ Đỗ Quỳnh Tiên", "Hoàng Hải Anh"])
tuoi = [18,18,18,18,18,17,18,19,20,19]
gioi_tinh = ["Nam","Nam","Nam","Nam","Nữ","Nam","Nữ","Nữ","Nữ","Nữ"]
```

```python
df2 = pd.DataFrame(np.array([ho_va_ten,tuoi,gioi_tinh]).T,columns = ["Họ và tên", "Tuổi", "Giới tính"])
df2
```

- Tạo dataframe như trên nhưng từ cấu trúc Dictionary có chứa các List giá trị tương ứng
với key. Xuất kết quả ra màn hình;

```python
df3 = pd.DataFrame({"Họ tên": ho_va_ten,"Tuổi": tuoi,"Giới tính":gioi_tinh})
df3
```
